```python
import tensorflow as tf
import numpy as np

loaded_data = np.loadtxt('../neowizard/MachineLearning/diabetes.csv',delimiter =',')

x_data = loaded_data[:,0:-1]
t_data = loaded_data[:,[-1]]

X = tf.placeholder(tf.float32 ,[None,8])
T = tf.placeholder(tf.float32 ,[None,1])

W = tf.Variable(tf.random_normal([8,1]))
b = tf.Variable(tf.random_normal([1]))

z = tf.matmul(X,W) +b

y = tf.sigmoid(z)

loss = - tf.reduce_mean(T * tf.log(y) + (1-T)*tf.log(1-y))

learning_rate = 0.05
optimizer = tf.train.GradientDescentOptimizer(learning_rate)
train = optimizer.minimize(loss)

predicted = tf.cast(y >0.5 , dtype =tf.float32)

accurancy = tf.reduce_mean(tf.cast(tf.equal(predicted,T),dtype = tf.float32))

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for step in range(20001):
        loss_val,_ =sess.run([loss,train],feed_dict={X:x_data ,T : t_data})
        
        if step % 500 == 0:
            print("step =",step,", loss_val =",loss_val)
            
        y_val,predict_val , accurancy_val = sess.run([y,predicted,accurancy],feed_dict={X:x_data ,T : t_data})
        
    print("\n y_val shape = ",y_val.shape ,", predicted_val =",predict_val.shape)
    print("\n Accurancy =",accurancy_val)
            

                           
```
